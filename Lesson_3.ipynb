### 4. Задание

```python
## 1
# Сначала построим график, чтобы определить приближенные решения
x = np.linspace(0.000001, 5, 201)
plt.plot(x, x**2 - 1, color='blue')
plt.plot(x, (np.exp(x) + x - 1)/x, color='red')
x2 = np.linspace(-5, -0.000001, 201)
plt.plot(x2, x2**2 - 1, color='blue')
plt.plot(x2, (np.exp(x2) + x2 - 1)/x2, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 30)
plt.grid(True)
plt.show()
```

```python
# На графике видно, что система имеет 3 решения
```

```python
def equations(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x*(1 - y) -1)
```

```python
x1, y1 = fsolve(equations, (4, 15))
print(x1, y1)
```

```python
x2, y2 = fsolve(equations, (2, 5))
print(x2, y2)
```

```python
x3, y3 = fsolve(equations, (-2, 3))
print(x3, y3)
```
