Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.

```python
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1
print(k, n, k/n)
c = np.math.factorial(4)/(np.math.factorial(2)*np.math.factorial(4-2))
P = c * (1/2**4)
print(P)
```

#### Задание 4

Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k

```python
for p in itertools.permutations("0123456", 5):
    print(''.join(x for x in p))
```

```python
for p in itertools.combinations("0123456", 5):
    print(''.join(p))
```

```python
for p in itertools.permutations("012", 3):
    print(''.join(x for x in p))
```

```python
for p in itertools.combinations("012", 3):
    print(''.join(p))
```
